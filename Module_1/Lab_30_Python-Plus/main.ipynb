# Challenge 2 - Building a composite indicator

In this part of the lab, we would build two different composite indicator, using some normalization functions.

Let's start by finding the number of rows where the CGPA is higher than 9 and the student has conducted research.

```python
# your code here
admissions.loc[(admissions["CGPA"] >= 9) & (admissions["GRE Score"] > 0)]
```

Now return all the rows where the CGPA is greater than 9 and the SOP score is less than 3.5. Find the mean chance of admit for these applicants.

```python
# your code here
not_so_admissible = admissions.loc[(admissions["CGPA"] >= 9) & (admissions["SOP"] > 0)]["Chance of Admit"].mean()
not_so_admissible
```

We would like to create a deciding factor column for each student. We standardize several columns and then pick the most important factor from a lookup table. If the standardized value is above 0.8, the student will be accepted.

We will start by creating a standardized column for `CGPA`, `GRE Score` and `LOR`. We will name these columns `CGPA_std`, `GRE_std`, and `LOR_std` respecively.

Recall that standardizing a column is done by subtracting the mean of the column from all observations in the column and then dividing each observation in the column by the column's standard deviation.

In the cell below, write a standardization function (a function that takes a column as input and returns a standardized column as output). Make sure to use the `numpy` versions of mean and standard deviation.

```python
def standardize(col):
    """
    This function takes a column from a dataframe 
    and returns a standardized column by subtracting the column's mean
    and dividing by the column's standard deviation.
    """
    return col.apply(lambda x: (x - col.mean()) / col.std())
```

Now create the standardized columns `CGPA_std`, `GRE_std`, and `LOR_std` and add them to the `admissions` dataframe.

```python
# your code here
admissions["GRE_std"] = standardize(admissions["GRE Score"])
admissions["CGPA_std"] = standardize(admissions["CGPA"])
admissions["LOR_std"] = standardize(admissions["LOR"])
admissions
```

Now we will try two different ways of making the decision whether or not to accept the student:
1. A weighted average of `CGPA_std`, `GRE_std`, and `LOR_std`.
2. The best score among `CGPA_std`, `GRE_std`, and `LOR_std`.

### Indicator 1.

Add a `score_1` column to the `admissions` dataframe, which is a weighted average, computed like this:

0.4 * `CGPA_std` + 0.3 `GRE_std` + 0.3 `LOR_std`

```python
# your code here
admissions["score_1"] = 0.4 * admissions["CGPA_std"] + 0.3 * admissions["GRE_std"] + 0.3 * admissions["LOR_std"]
admissions
```

### Indicator 2.

Add a `score_2` column to the `admissions` dataframe, which is the best (highest) score among `CGPA_std`, `GRE_std`, and `LOR_std`.

```python
# your code here
admissions["score_2"] = admissions[["CGPA_std", "GRE_std", "LOR_std"]].max(axis=1)
admissions
```

### Decision

Create a column called `decision_1` with the admission decision based on score 1: `True` if `score_1` is greater than 0.8 and `False` otherwise.

Similarty, create a `decision_2` column that contains the decision for score 2.

```python
# your code here
admissions["decision_1"] = admissions["score_1"].apply(lambda x: x > 0.8)
admissions["decision_2"] = admissions["score_2"].apply(lambda x: x > 0.8)
admissions
```

How many applicants will be accepted to the program using each scoring method?

```python
# your code here
accepted_1 = admissions.loc[admissions["decision_1"] == True]["index"].count()
accepted_2 = admissions.loc[admissions["decision_2"] == True]["index"].count()
accepted_1, accepted_2
```
