```python
import pandas as pd
from sklearn.impute import KNNImputer
import numpy as np
# Read the data
df = pd.read_excel('./example_data_cleaning.xlsx')

# Check the data
df.head()
```

```python
# Drop two last columns and the last row:
df.drop(["Unnamed: 7", "Unnamed: 8"], axis=1, inplace=True)
df.drop(df.tail(1).index, inplace=True)
df
```

```python
# Impute missing values in the column "BirthYear"
imputer = KNNImputer(n_neighbors=5)
df["BirthYear"] = imputer.fit_transform(df["BirthYear"].values.reshape(-1, 1))
df

```

```python
# Convert all float columns to integers
numerical = df.select_dtypes(include=np.number)
df[numerical.columns] = numerical.astype(int)
df
```

```python
# Export dataframe as csv:
df.to_csv("example_data_cleaning.csv", index=False)
```
