# StackAPI

#### Import the necessary libraries here:

```python
from stackapi import StackAPI
from datetime import datetime
```

#### Question 1: Find the questions and answers of last month.

```python
site = StackAPI('stackoverflow')
lastMonthEpoch = int(datetime(2023, 3, 18).timestamp())
todayEpoch = int(datetime(2023, 4, 18).timestamp())
questions = site.fetch('questions', fromdate=lastMonthEpoch, todate=todayEpoch)['items']
answers = site.fetch('answers', fromdate=lastMonthEpoch, todate=todayEpoch)['items']
print(questions, answers)
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
yesterday = datetime(2023, 4, 17)
mostVoted = site.fetch('questions', fromdate=yesterday, tagged='python', sort='votes', max=1)['items']
print(mostVoted)
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
answer6784 = site.fetch('answers', ids=[6784])['items']
answer6473 = site.fetch('answers', ids=[6473])['items'] # this answer does not exist!
print(answer6784, answer6473)
```
